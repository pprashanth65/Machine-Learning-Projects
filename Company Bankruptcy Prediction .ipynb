```python
from scipy.stats import spearmanr
```

```python
cols=[]
for i in df.columns:
    if i!='Bankrupt':
        result=spearmanr(df['Bankrupt'],df[i])
        if result.pvalue<0.05:
            print(f"The column name is {i}\n")
            print(f"We are accepting Null hypothesis and the column is not significant\n")
            print("-------------------------------------------------------------------")
        else:
            print(f"The column name is {i}\n")
            print(f"We are rejecting Null hypothesis and accepting Alternate hypothesis and the column is significant\n")
            print("-------------------------------------------------------------------")
            cols.append(i)
```

```python
from pyod.models.abod import ABOD
from pyod.models.knn import KNN
from pyod.models.cblof import CBLOF
from pyod.models.pca import PCA
from pyod.models.loda import LODA
from pyod.models.iforest import IsolationForest
from pyod.models.hbos import HBOS
from pyod.models.sod import SOD
from pyod.models.ocsvm import OCSVM
```

```python
def outliers(x):
    a={}
    algorithms = [ABOD(), KNN(n_jobs=-1),CBLOF(),PCA(),LODA(),HBOS(),SOD(),OCSVM()]
    for algorithm in algorithms:
            pred=algorithm.fit_predict(x)
            a[algorithm]=pred
    return a 
            
```
